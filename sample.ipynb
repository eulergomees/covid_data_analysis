```python
import numpy as np
import pandas as pd
from urllib.request import urlretrieve
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv '
urlretrieve(url, 'datasets/global_covid_cases.csv')

df_covid = pd.read_csv('datasets/global_covid_cases.csv')

#print(df_covid)

df_covid.drop(['Lat', 'Long'], axis=1, inplace=True) #axis = 1 is column/inplace = true overwrite the original dataframe

df_country = df_covid.groupby('Country/Region').sum()

s_brazil = df_country.loc['Brazil']
s_brazil = s_brazil[s_brazil > 0]

#plot
plt.figure(figsize=(15,8))
plt.bar(s_brazil.index, s_brazil.values)
```

```python
tam = len(s_brazil)
plt.figure(figsize=(14,8))
plt.xticks(rotation=60)
plt.bar(s_brazil.index[tam-50:tam], s_brazil.values[tam-50:tam])
plt.title('Covid Cases')
```
